```python
import kagglehub
import os
import pandas as pd

# 1. Download (or reuse cached) dataset
path = kagglehub.dataset_download("daryaheyko/airline-on-time-statistics-and-delay-causes-bts")
print("Path to dataset files:", path)

# 2. See what files are there
print(os.listdir(path))
```

```python
csv_path = os.path.join(path, "Airline_Delay_Cause.csv")  # adjust name if different
df = pd.read_csv(csv_path)

df.head()

```

```python
from pymongo import MongoClient

# connect to local MongoDB
client = MongoClient("mongodb://localhost:27017/")

# create database + collection
db = client["airline_delay_db"]
collection = db["airline_delay_cause"]

# convert dataframe to MongoDB-dict format
data = df.to_dict("records")

# insert
collection.insert_many(data)

print("Imported into MongoDB successfully!")
```
