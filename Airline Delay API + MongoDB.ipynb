```python
import pandas as pd

path = "/Users/shirleyxpy/.cache/kagglehub/datasets/daryaheyko/airline-on-time-statistics-and-delay-causes-bts/versions/2"

csv_path = f"{path}/Airline_Delay_Cause.csv"
df = pd.read_csv(csv_path)

df.head()
```

```python
from pymongo import MongoClient

# connect to local MongoDB
client = MongoClient("mongodb://localhost:27017/")

# create database + collection
db = client["airline_delay_db"]
collection = db["airline_delay_cause"]

# convert dataframe to MongoDB-dict format
data = df.to_dict("records")

# insert
collection.insert_many(data)

print("Imported into MongoDB successfully!")
```
